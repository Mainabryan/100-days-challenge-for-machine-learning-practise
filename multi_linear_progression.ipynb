# IMPORTING THE FILES

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Prepare the data

```python
# define the independent atributes(house size,bedrooms,house age)
house_features = np.array([
        [50, 2, 10],
    [75, 3, 5],
    [100, 3, 8],
    [125, 4, 4],
    [150, 4, 2]
])
# define the dependenat vaiable(house price in $)
house_prices = np.array([150000, 200000, 250000, 300000, 350000])
```

# train the model

```python
# initialize the lineaar regression
model =LinearRegression()
model.fit(house_features,house_prices)
```

# Retrieve the Learned Equation

```python
# Get coefficients and intercept
m_values = model.coef_  # Slopes
b = model.intercept_  # Intercept
# Display the equation
print(f"Equation of the regression line: Y = {m_values[0]:.2f}X1 + {m_values[1]:.2f}X2 + {m_values[2]:.2f}X3 + {b:.2f}")
```

# MAKE A PREDICTION

```python
# NEW HOUSE DATA FOR PREDICTION
new_house = np.array([[120,3,6]])
predicted_price = model.predict(new_house)
print(f"Predicted price for the new house: ${predicted_price[0]:.2f}")
```

# plot the graph

```python
plt.scatter(house_features[:, 0], house_prices, label='Actual Prices', color='blue')
plt.scatter(new_house[0, 0], predicted_price, label='Predicted Price', color='red')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('House Price ($)')
plt.title('House Price Prediction')
plt.legend()
plt.show()
```
