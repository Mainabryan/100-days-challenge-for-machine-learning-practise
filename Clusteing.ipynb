```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("/content/Mall_Customers (1).csv")

# Drop CustomerID and convert Gender to numerical
data = data.drop("CustomerID", axis=1)
data["Gender"] = data["Gender"].map({"Male": 1, "Female": 0})

# Fit Agglomerative Clustering model (use 'metric' instead of 'affinity')
model = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='complete')
clust_labels = model.fit_predict(data)

# Optional: Add cluster labels back to the dataframe for inspection or use
data["Cluster"] = clust_labels

# Visualization
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
scatter = ax.scatter(data['Annual Income (k$)'],
                     data["Spending Score (1-100)"],
                     c=data["Cluster"],
                     s=50,
                     cmap='viridis')
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("Annual Income (k$)")
ax.set_ylabel("Spending Score (1-100)")
plt.colorbar(scatter)
plt.show()
```
