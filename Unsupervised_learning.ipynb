## CLUSTERING

# Import the necessary libraies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
```

# Load the small dataset

```python
# Creating a small dataset of 2D points
data = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [8, 7],
    [8, 8],
    [25, 80]
])

# Optional: Turn into a DataFrame for easy viewing
df = pd.DataFrame(data, columns=['X', 'Y'])
print(df)
```

#Visualize the Points

```python
plt.scatter(df['X'], df['Y'], color='blue')
plt.title("Original Data Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

#Create the Dendrogram

```python
# Create linkage matrix using Wardâ€™s method
linked = linkage(data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Dendrogram (Hierarchical Clustering)")
plt.xlabel("Data Point Index")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

#Appply agglomarative clustering

```python
model = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = model.fit_predict(data)

# Plot the clustered data
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow')
plt.title("Hierarchical Clustering Results (2 Clusters)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
