```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting

# Study hours (X1)
study_hours = np.array([2, 3, 4, 5, 6, 7, 8])

# Other features: [sleep_hours, practice_tests, final_score]
students_performance = np.array([
    [5, 1, 60],
    [6, 2, 65],
    [4, 1, 68],
    [5, 3, 75],
    [6, 2, 80],
    [5, 4, 85],
    [6, 5, 90]
])

# Extract X2 and X3
X2_X3 = students_performance[:, :2]  # sleep_hours, practice_tests
Y = students_performance[:, 2]       # final_score

# Combine study_hours (X1) with sleep_hours (X2) and practice_tests (X3)
X = np.column_stack((study_hours, X2_X3))

# Train model
model = LinearRegression()
model.fit(X, Y)

# 3D Scatter Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Extract X1, X2, X3 for plotting
X1 = X[:, 0]
X2 = X[:, 1]
X3 = X[:, 2]

# Plot points in 3D
sc = ax.scatter(X1, X2, Y, c='red', label='Final Score', s=60)

# Label axes
ax.set_xlabel("Study Hours (X1)")
ax.set_ylabel("Sleep Hours (X2)")
ax.set_zlabel("Final Score (Y)")
ax.set_title("3D Scatter: Study Hours, Sleep Hours vs Final Score")

plt.legend()
plt.show()
```
