# import the files

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from sklearn.linear_model import LinearRegression
```

# PREPARING THE DATA

```python
# DEFINE THE DATASET (SLEEP_HOURS, PRACTICE_TESTS, FINAL_SCORE)
students_performance = np.array([
    [5, 1, 60],
    [6, 2, 65],
    [4, 1, 68],
    [5, 3, 75],
    [6, 2, 80],
    [5, 4, 85],
    [6, 5, 90]
])

# Define the variable: study_hours
study_hours = np.array([2, 3, 4, 5, 6, 7, 8])

```

# Train the model

```python
# initialize the model
model = LinearRegression()
model.fit(students_performance,study_hours)
```

# Retrive the dataeet

```python
# get the coefficeienta nd the intercepts
m_values = model.coef_  # Changed m_valuesn to m_values
b = model.intercept_


print(f"Equation of the regression line: Y = {m_values[0]:.2f}X1 + {m_values[1]:.2f}X2 + {m_values[2]:.2f}X3 + {b:.2f}")
```

# MAKE TE PREDICTION

```python
new_performance = np.array([[6,5,3]])
predicted_study_hours = model.predict(new_performance)
print(f"Predicted study hours: {predicted_study_hours[0]:.2f}")
```

# plot the dataset

```python
# plot the scatter
plt.scatter(students_performance[:,0],study_hours,color='blue',label='data points')
plt.scatter(new_performance[:,0],predicted_study_hours,color='red',label='predicted points')
plt.xlabel('students_performance')
plt.ylabel('study_hours')
plt.legend()
plt.show()
```
